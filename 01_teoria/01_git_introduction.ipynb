# Introduzione a git

Git è il più utilizzato sistema di *code versioning*, cioè uno strumento per tenere traccia di come i file di un progetto software sono stati modificati nel corso della realizzazione. Durante la fase di produzione capita spesso di dover modificare il codice su cui si sta lavorando o di lasciare in sospeso il lavoro per un certo lasso di tempo. I sistemi di code versioning offrono un modo per memorizzare quali sono state le modifiche apportate ai singoli file e quindi ripercorrere in qualunque momento la loro storia e perfino di ripristinare versioni precedenti.

Imparare ad utilizzare bene git rappresenta uno step necessario per qualsiasi programmatore, questo notebook rappresenta una guida per imparare a muovere i primi passi.

## Installare git

Il primo passo per utilizzare git è installare il software liberamente disponibile al seguente link: [https://git-scm.com/](https://git-scm.com/)

Una volta installato è possibile verificare il corretto funzionamento semplicemente aprendo un terminale e provando a controllare la versione di git installata:

```bash
git --version
```

La prima volta che si installa git è buona prassi configurare le variabili globali come utente e contatto di riferimento.

Per verificare la configurazione attuale è sufficiente scrivere nel terminale:
```bash
git config --global --list
```

Solitamente si configurano almeno.
- nome `git config --global user.name "[name]"`
- mail `git config --global user.email "[email address]"`

Opzionale per una visualizzazione migliore: `git config --global color.ui auto`

## Inizializzare una repository

Per inizializzare un repository, cioè iniziare a tenere traccia delle modifiche apportate ai file in esso contenuti, è sufficiente scrivere nel terminale la prima volta il seguente comando:

```bash
git init
```

Questo crea una cartella `.git` nella quale si inizia a tracciare i file del repository. Questa cartella non è visualizzata di default in VS Code, per renderla visibile è sufficiente aprire le `impostazioni`, cercare la voce `Files: Exclude`, rimuovere la voce `.git`.

Nota Bene: la cartella contiene la storia del repository, nel caso sia eliminata la storia va persa.

## Commits

A questo punto si può iniziare ad aggiungere e modificare i file contenuti nel repository per realizzare il proprio progetto.
in ogni momento è possibile  visualizzare lo stato delle modifiche scrivendo nel terminale:
```bash
git status -s
```

Normalmente si modificano più file e a un certo punto si decide di salvare le modifiche nella storia del progetto. Digitando `git status -s` comparirà l'elenco di tutti i file modificati con associata una lettera descrive il tipo di modifica. A questo punto si sceglie in genere di salvare le modifiche a gruppetti in base al motivo per cui sono state apportate: ad esempio posso aver modificato i file ex_1.py e utils.py per rendere il codice più leggibile. Per alvare questo gruppo di mofiche prima vanno aggiunte con il comando `git add`; esempio

```bash
git add ex_1.py
git add utils.py
```

Poi va effettuato il commit che registra le modifiche

```bash
git commit -m "Modified ex_1.py: ..."
```

VS Code offre un'interfccia grafica per effettuare con semplicità questa operazione. Basta aprire la scheda `controllo del codice sorgente`. Qui le modifiche sono elencate sotto l'omonima voce. Per aggiungerle basta premere il pulsante `+`, poi si inserisce un messaggio e in seguito si preme sul pulsante `Commit`.


Per visualizzare la storia dei commits si digita nel terminale:
```bash
git log --oneline
```

VS Code presenta anche una tendina nominata `GRAFO` che permette di visuallizare la storia dei commit che sono stati effettuati.
